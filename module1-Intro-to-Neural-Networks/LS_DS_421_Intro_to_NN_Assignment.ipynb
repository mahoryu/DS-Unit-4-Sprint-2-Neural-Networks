## Define the Following:
You can add image, diagrams, whatever you need to ensure that you understand the concepts below.

### Input Layer:
- This is the inpout from the dataset and is sometimes called the visible layer.

### Hidden Layer:
- The layer or layers that come after the input layer and the user has no direct interaction with them. 
  They perform their functions and then pass it onto either the next hidden layer or the output layer.
 
### Output Layer:
- This is the final layer that ouputs a vector of values that show what the actual prediction for the function is.

### Neuron:
- Or the nodes are what receive inputs from the previous layer or the dataset and then give outputs for either 
  the entire function or for the next layer to use.

### Weight:
- The weight show how important the output for each node plays in determining the output of the function.

### Activation Function:
- The function that modifies the output of the function so that it is understandable to humans. 
  These can also be used to convert the output of one layer to be usable by the next hidden layer.

### Node Map:
- The Node Map is a visual diagram that shows how the neural network works. How the outputs of of node move on to 
  the next node so that you can have an idea of how it works.

### Perceptron:
- The simplest king of neural network that only has a single node that can take any number of inputs and spit out 
  an output.

## Write your own perceptron code that can correctly classify (99.0% accuracy) a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
import pandas as pd
import numpy as np
np.random.seed(777)

data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0]
       }

df = pd.DataFrame.from_dict(data).astype('int')
df.values
```

```python
##### Your Code Here #####

# define the sigmoid and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
# Initialize the random weights for the inputs
weights = 2 * np.random.random((2,1)) - 1
weights
```

```python
# separate the inputs from the output
inputs = df[["x1","x2"]]
correct_out = [[1],[1],[1],[0]]
```

```python
# create a for loop to update our weights 1,000 times
for iteration in range(1000):
    # weighted sum of inputs / weights
    weighted_sum = np.dot(inputs, weights) + 7
    # activate
    activated_output = sigmoid(weighted_sum)
    # calculate error
    error = correct_out - activated_output
    adjustments = error * sigmoid_derivative(weighted_sum)
    # Update the weights
    weights += np.dot(inputs.T, adjustments)
#     breakpoint()
    
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```

Although neural networks can handle non-normalized data, scaling or normalizing your data will improve your neural network's learning speed. Try to apply the sklearn `MinMaxScaler` or `Normalizer` to your diabetes dataset. 

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer

feats = list(diabetes)[:-1]

X = diabetes[feats]
y = diabetes['Outcome']
transformer = Normalizer().fit(X)
X = pd.DataFrame(transformer.transform(X))
X.columns = feats
X.head()
```

```python
y = np.array(y).reshape(768,1)
```

```python
##### Update this Class #####

class Perceptron:
    
    def __init__(self, niter = 10):
        self.niter = niter
    
    def __sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        sx = sigmoid(x)
        return sx * (1 - sx)

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """
        
        # Randomly Initialize Weights
        weights = 2 * np.random.random((len(X.columns),1)) - 1

        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, weights) + 11
            # Activate!
            activated_output = sigmoid(weighted_sum)
            # Cac error
            error = y - activated_output
            adjustments = error * sigmoid_derivative(weighted_sum)
            # Update the Weights
            weights += np.dot(X.T, adjustments)
            
        self.weights = weights
        self.activated_output = activated_output

    def predict(self, X):
        """Return class label after unit step"""
        print(self.activated_output)
        
```

```python
perceptron = Perceptron(200)
perceptron.fit(X,y)
perceptron.predict(X)
```
