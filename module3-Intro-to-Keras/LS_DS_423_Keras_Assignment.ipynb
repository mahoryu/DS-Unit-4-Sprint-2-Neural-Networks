### Import the Boston Dataset

```python
##### Your Code Here #####
import pandas as pd

from tensorflow.keras.datasets import boston_housing
(X_train, y_train), (X_test, y_test) = boston_housing.load_data(path="boston_housing.npz", test_split=0.2, seed=777)
```

```python
headers = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
target = ['MEDV']
```

```python
X_train = pd.DataFrame(X_train, columns=headers)
X_test = pd.DataFrame(X_test, columns=headers)
y_train = pd.DataFrame(y_train, columns=target)
y_test = pd.DataFrame(y_test, columns=target)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train.head()
```

### Normalize the features

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer

def normalization(Xtrain, Xtest):
    transformer = Normalizer().fit(Xtrain)
    Xtrain = pd.DataFrame(transformer.transform(Xtrain))
    Xtrain.columns = headers
    Xtest = pd.DataFrame(transformer.transform(Xtest))
    Xtest.columns = headers
    return Xtrain,Xtest
    
X_train, X_test = normalization(X_train,X_test)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#import Keras models needed
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

### Instantiate a model object and add layers to your model

```python
model = Sequential()
model.add(Dense(10, input_dim=13))
model.add(Dense(1))

callback = EarlyStopping(patience=3, verbose=1, min_delta=0.01)
```

```python
# comple it
model.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
model.summary()
```

```python
model.fit(X_train,y_train,
         epochs=100,
         validation_data=(X_test,y_test),
         callbacks=[callback])
```
